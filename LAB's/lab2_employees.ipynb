You need to create a model that will accurately predict which of the employees can **leave**.

## 1. Data Cleaning and Preparation

```python
import numpy as np
import pandas as pd 
df = pd.read_csv('empl_train.csv')
df_test = pd.read_csv('empl_test.csv')
df
```

```python
df['Gender'].unique()
```

```python
df = df.drop(columns=(['Education', "City","JoiningYear"]))
df["Gender"].replace({"Female":0, "Male":1}, inplace=True)
df["EverBenched"].replace({"No":0, "Yes":1}, inplace=True)
df["Gender"].replace({"Женский":0, "Мужской":1}, inplace=True)
df = pd.get_dummies(df, columns=["PaymentTier"])
df = pd.get_dummies(df, columns=["ExperienceInCurrentDomain"])

df.head()
```

```python
X = df.copy()
y = X.pop("Leave")
```

```python
from sklearn.model_selection import train_test_split as tts
X_train, X_test, y_train, y_test = tts(X, y, test_size=0.2)
```

## 2. Modeling (ML)
Build and compare performance of kNN, Logistic Regression, Decision Tree Classifier and Random Forest Classifier models.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
from sklearn.ensemble import RandomForestClassifier

#LOGISTICREGRESSION

logi = LogisticRegression()
logi.fit(X_train,y_train)
logis = logi.score(X_train,y_train)
print("LOGISTIC REGRESSION:",logis)

#DECISIONTREE

dec = DecisionTreeClassifier(max_depth =9, random_state=42)
dec.fit(X_train, y_train)
dectree = dec.score(X_train, y_train)
print("DECITIONTREE:", dectree)

#RANDOMFOREST
rf = RandomForestClassifier(n_estimators = 100)
rf.fit(X_train,y_train)
randfor = rf.score(X_train,y_train)
print("RANDOM FOREST:", randfor)
```

## TEST DATA PREPARATION

```python
df_test = df_test.drop(columns=(['Education', "City","JoiningYear"]))
df_test["Gender"].replace({"Female":0, "Male":1}, inplace=True)
df_test["EverBenched"].replace({"No":0, "Yes":1}, inplace=True)
df_test["Gender"].replace({"Женский":0, "Мужской":1}, inplace=True)
df_test = pd.get_dummies(df_test, columns=["PaymentTier"])
df_test = pd.get_dummies(df_test, columns=["ExperienceInCurrentDomain"])
```
