## kNN in sklearn

Try to check the score of KNN algorithm on test data.

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
X_train = df[['PetalLengthCm', 'PetalWidthCm']]
y_train = df['Species']
X_test = df2[['PetalLengthCm', 'PetalWidthCm']]

knn = KNeighborsClassifier(n_neighbors=5)

knn.fit(X_train, y_train)

y_test_pred = knn.predict(X_test)

plt.scatter(df['PetalLengthCm'], df['PetalWidthCm'], c=y_train)
plt.scatter(df2['PetalLengthCm'], df2['PetalWidthCm'], c=y_test_pred, marker='x')

plt.legend()
plt.show()
```
