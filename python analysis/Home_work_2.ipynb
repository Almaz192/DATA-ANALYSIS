## Теория

В Seaborn встроены функции для визуализации распределения данных displot в следующих форматах:

    histplot
    kdeplot
   

Функция histplot() в Seaborn используется для построения гистограммы, которая показывает распределение данных. Она группирует значения в "бины" (столбики) и отображает частоту (или плотность) значений внутри каждого интервала. Функция принимает следующие параметры:

   - data - Массив данных или колонка в DataFrame
   - bins - Количество бинов (столбиков) для группировки данных
   - kde - Если True, добавляет график плотности распределения (KDE)
   - stat - Может быть "count", "density" или "percent" (Использует нормированную плотность вместо абсолютных частот)
   - hue - Разделение данных по категориям (например, пол)
   - element -	Варианты "bars" (столбцы по умолчанию), "step" (контур) или "poly" (заполненный контур)
   - multiple - "layer" (по умолчанию), "stack" (сложенные гистограммы), "dodge" (раздельные столбцы), "fill" (как 100% stacked)
   - discrete - Если True, рассматривает данные как дискретные
   - cumulative - Если True, строит накопительную гистограмму (Каждое значение включает все предыдущие, создавая эффект накопления)

  

```python
#пример
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
sns.histplot(tips, x="total_bill", bins=20, kde=True, hue='sex') 
```

## Теория

Другими словами, histplot() рисует гистограмму — это график, который показывает, как часто встречаются разные значения в данных.

Как это работает?
- Берет все числа из списка (например, цены заказов в кафе).
- Делит их на группы (например, от 0 до 10, от 10 до 20 и т. д.).
- Считает, сколько чисел попадает в каждую группу.
- Рисует столбики: чем выше столбик, тем больше значений в этой группе.

## Теория

kdeplot() в Seaborn используется для построения графика плотности вероятности (KDE — Kernel Density Estimate), который помогает оценить, как распределены данные, используя гладкую кривую. Он показывает плотность данных — как часто встречаются определенные значения в наборе данных. Вместо того чтобы использовать прямоугольники для подсчета частоты, KDE использует ядра (обычно гауссовые функции), чтобы оценить плотность данных в каждой точке. Результатом является плавная кривая, которая показывает вероятность появления значений в различных диапазонах. Функция принимает следующие параметры:

   - data - Массив данных или колонка в DataFrame
   - fill - Если True, график будет заполнен цветом (заполняет область под кривой)
   - color - Цвет линии графика
   - bw - Параметр для контроля ширины сглаживания (чем меньше значение, тем менее сглаженная линия)
   - hue - Добавляет разделение по категориям (например, для разных групп)


```python
# пример
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

sns.kdeplot(tips["total_bill"], fill=True)
```

## (1 балл) Задание 5

Разбейте возраст пассажиров Титаника на 4 возрастные группы (age_group):

Дети: 0-12 лет
Молодые: 13-25 лет
Взрослые: 26-50 лет
Пожилые: 51+ лет

Постройте график, который отображает уровень выживаемости (survived) для каждой возрастной группы (age_group). Рассчитайте среднее значение выживаемости в каждой группе.

```python
# ваш ответ
bins = [0, 12, 25, 50, np.inf]
labels = ["Дети", "Молодые", "Взрослые", "Пожилые"]
df["age_group"] = pd.cut(df["age"], bins=bins, labels=labels)

survival_by_age_group = df.groupby("age_group")["survived"].mean()

fig, ax = plt.subplots(figsize=(8, 6))
sns.barplot(x=survival_by_age_group.index, y=survival_by_age_group.values, ax=ax)

ax.set_title("Выживаемость пассажиров по возрастным группам")
ax.set_xlabel("Возрастная группа")
ax.set_ylabel("Средняя доля выживших")
```
