** Описание датасета: **

•	Country – Страна, где был снят фильм/сериал

•	Data_added – Дата добавления на Netflix

•	Director – Режиссер фильма/сериала

•	Duration – Общая продолжительность (в минутах или количестве сезонов)

•	Listed_in – Жанры

•	Rating  – Рейтинг фильма/сериала на ТВ

•	Release_year – Фактический год выпуска фильма/сериала

•	Show_id – Уникальный идентификатор для каждого фильма/сериала

•	Title – Название фильма/сериала

•	Type – тип фильм или сериал


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df=pd.read_csv("netflix_titles.csv")
df.head()
```

## (1 балл) Задание 1

## Постройте столбчатую диаграмму, которая отображает количество различных категорий шоу на платформе Netflix. На оси X укажите типы шоу, такие как фильмы/сериалы, а на оси Y — количество шоу в каждой из этих категорий.
** Добавьте заголовок диаграмме и закрасьте столбцы.


Для закрашивания столбцов используйте уникальные цвета Netflix 
#221f1f
#b20710
#e50914
#f5f5f1

```python
count = df['type'].value_counts()
```

```python
# ваш код
fig, ax = plt.subplots()
count.plot(kind = "bar", color=['#221f1f', '#b20710', '#e50914', '#f5f5f1'], width=0.6, edgecolor='black')
ax.set_title('Bar Plot')
ax.set_xlabel('Количество различных категорий шоу на платформе Netflix')
ax.set_ylabel('Значения')
plt.show()
```

## (1 балл) Задание 2

## Постройте столбчатую диаграмму со 100% накоплением, которая отображает количество различных категорий шоу на платформе Netflix. 
** Добавьте заголовок диаграмме и закрасьте столбцы.

```python
# ваш код

type_counts = df['type'].value_counts(normalize=True) * 100  
fig, ax = plt.subplots(figsize=(6, 4))
type_counts.plot(kind='bar', stacked=True, color=['#221f1f', '#e50914'], ax=ax)
ax.set_ylabel("Percentage")
ax.set_xlabel("Show Type")
ax.set_title("Distribution of Show Types on Netflix (100% Stacked)")
ax.set_xticklabels(type_counts.index, rotation=0)

plt.show()
```

## (1 балл) Задание 3

## Объедините графики с упражнения 1 и упражнения 2 в рамках одной сетки. 

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

count = df['type'].value_counts()
count.plot(kind="bar", color=['#221f1f', '#b20710'], width=0.6, edgecolor='black', ax=axes[0])
axes[0].set_title('Количество различных категорий шоу на Netflix')
axes[0].set_xlabel('Тип шоу')
axes[0].set_ylabel('Количество')

type_counts = df['type'].value_counts(normalize=True) * 100
type_counts.plot(kind='bar', stacked=True, color=['#221f1f', '#e50914'], ax=axes[1])
axes[1].set_ylabel("Percentage")
axes[1].set_xlabel("Тип шоу")
axes[1].set_title("Распределение категорий шоу (100% Stacked)")

# Display the combined plots
plt.tight_layout()
plt.show()
```
