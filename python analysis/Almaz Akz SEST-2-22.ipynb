```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('health_tracker.csv')
df
```

```python
grouped_data = df.groupby('Activity_Type')['Calories_Burned'].sum().sort_values(ascending=False)
fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(grouped_data.index, grouped_data.values, color=['#363537','#EF2D56','#ED7D3A','#8CD867','#2FBF71'])
ax.set_title('Общее количество сожжённых калорий по типу активности', fontsize=14, fontweight='bold')
ax.set_xlabel('Тип активности', fontsize=12)
ax.set_ylabel('Сожжённые калории', fontsize=12)
ax.grid(True, linestyle=':', alpha=0.6)
plt.xticks(rotation=45, fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.show()
```

```python
sns.set_context('paper')
sns.set_style('white')

sleep_quality_counts = df['Sleep_Quality'].value_counts().reset_index()
sleep_quality_counts.columns = ['Sleep_Quality', 'Count']

plt.figure(figsize=(8, 5))
ax = sns.barplot(x='Sleep_Quality', y='Count', data=sleep_quality_counts, palette='rocket')

ax.set_title('Количество записей по качеству сна', fontsize=14, fontweight='bold')
ax.set_xlabel('Качество сна', fontsize=12)
ax.set_ylabel('Количество записей', fontsize=12)
ax.grid(True, linestyle=':', alpha=0.6)


plt.tight_layout()
plt.show()
```

```python

sns.set_context('notebook')
sns.set_style('darkgrid')

plt.figure(figsize=(10, 6))
ax = sns.histplot(data=df, x='Duration_Minutes', hue='Activity_Type', multiple='stack', bins=20)

ax.set_title('Распределение продолжительности активности по типу', fontsize=14, fontweight='bold')
ax.set_xlabel('Продолжительность (мин)', fontsize=14, fontfamily='monospace', fontstyle='italic')
ax.set_ylabel('Частота', fontsize=14, fontfamily='monospace', fontstyle='italic')

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_color('yellow')
ax.spines['left'].set_linewidth(2)
ax.spines['bottom'].set_color('blue')
ax.spines['bottom'].set_linewidth(2)

plt.tight_layout()
plt.show()
```
