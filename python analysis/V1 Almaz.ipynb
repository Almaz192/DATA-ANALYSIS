```python
df = df.drop("_ProductID", axis=1)
```

```python
print(df.columns)
```

```python
df['Unit Profit'] = df['Unit Price']*(1-df['Discount Applied'])-df['Unit Cost']
df
```

```python
df['Total Profit'] = df['Unit Profit'] * df['Order Quantity']
df
```

```python
df['Sum of Sales'] = df['Unit Price'] * (1 - df['Discount Applied']) * df['Order Quantity']
```

```python
sns.set_style('whitegrid')
df['Sales Team'].value_counts().head().plot(kind = 'barh',title = 'Заказы которые приходится на каждого продавца ')
```

```python
a = df.groupby('Sales Channel')['Sum of Sales'].describe()
a
```

```python
o = df['Order Quantity'].describe()
o
```

```python
p = df.groupby('Product name')['Order Quantity'].sum().head(10).sort_values(ascending=False).plot(kind = 'barh',title='Cумма проданных товаров по каждому типу товара')
p
```

```python
pivot_table = df.pivot_table(values='Sum of Sales', index='Sales Team', columns='Product name', aggfunc='sum', fill_value=0)
pivot_table
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15, 8)) 
sns.heatmap(pivot_table, cmap='coolwarm', annot=False, fmt=".0f", linewidths=0.5) 
plt.title('Сумма продаж по командам продаж и товарам', fontsize=14) 
plt.xlabel('Product name', fontsize=12)
plt.ylabel('Sales Team', fontsize=12)
plt.xticks(rotation=45, ha='right') 
plt.yticks(rotation=0)  

plt.show()
```
