```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
list_num_of_rooms =[]
list_area=[]
list_floor=[]
list_address=[]
list_price_usd=[]
list_price_som=[]
list_out_of=[]
```

```python
list_num_of_rooms
```

```python
for i in range(1,11):
    html_page=requests.get('https://www.house.kg/kupit-kvartiru?page={i}').content
    soup= BeautifulSoup(html_page,"lxml")
    vse_kvar=soup.find_all("div",class_="listing")
    for kvartira in vse_kvar:
        title = kvartira.find("p",class_="title").text.strip()
        title= title.split(",")
        price_usd = kvartira.find("div",class_="price").text
        price_som = kvartira.find("div",class_="price-addition").text
        address = kvartira.find("div",class_="address").text.strip()
        title_2 = title[2].replace(" этаж из "," / ")
        title_2 = title_2.split("/")
        list_area.append(float(title[1].replace(" м2","")))
        list_num_of_rooms.append(title[0].replace("-комн. кв.",""))
        list_floor.append(title_2[0])
        list_out_of.append(title_2[1])
        list_address.append(address)
        list_price_usd.append(int(price_usd.replace(" ","").replace("$","")))
        list_price_som.append(int(price_som.replace(" ","").replace("сом","")))
```

```python
data = pd.DataFrame.from_dict(
    {'number of rooms': list_num_of_rooms, 
        'area (in m2)': list_area,
        'floor': list_floor,
         'out of floor' : list_out_of,
        'address': list_address,
        'price in USD': list_price_usd,
        'price in KGS': list_price_som

    })
```
