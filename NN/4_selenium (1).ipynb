```python
from selenium import webdriver
```

```python
import time
```

```python
url = 'https://globus-online.kg/catalog/kulinariya/'

# Загрузка драйвера
driver = webdriver.Chrome()

driver.get(url)

time.sleep(3)

SCROLL_PAUSE_TIME = 3

# Расчет высоты страницы
last_height = driver.execute_script("return document.body.scrollHeight")

while True:
    # Scroll down to bottom
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

    # Подождать загрузки страницы
    time.sleep(SCROLL_PAUSE_TIME)

    # Расчет новой высоты и сравнение с предыдущей. Если разницы нет, то страница дошла до конца
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height

html_text = driver.page_source
time.sleep(1) # чтобы в конце не сразу закрывалось
driver.quit()
```

```python
soup= BeautifulSoup(html_text,"lxml")
```

```python
# Название
# Цена
# Единица измерения (кг, шт)
all_products = soup.select("div[class*=js-element]") # with *= means: contains
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
list_names =[]
list_price=[]
list_demension=[]

for i in all_products:
    title = i.find("div",class_="list-showcase__name").text.strip()
    list_names.append(title)
    price = i.find("span", class_="c-prices__value js-prices_pdv_ГЛОБУС Розничная").text.strip().split("сом/")
    list_price.append(price[0])
    list_demension.append(price[1])
```
